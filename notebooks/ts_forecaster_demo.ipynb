## Example data

## Import the models

```python
# imports lstm, var, cnn, gru, sarima, xgboost

from src.models.lstm import lstm
from src.models.cnn import cnn
from src.models.gru import gru
from src.models.sarima import sarima
from src.models.XGBoost import xgboost
from src.models.var import var
```

## Implement and use models

### LSTM

```python
# Specify your own parameters 
test_size_prct = 0.3
target = "cpu"
max_lag = 5

# initialise the model
lstm_model = lstm(df = df, target = target, test_size = test_size_prct, plot_learning_curves_bool = True)

# train and test the model accordingly
lstm_forecast, test, train = lstm_model.fit()
```

### CNN

```python
# initialise the model
cnn_model = cnn(df = df, target = target, test_size = test_size_prct, plot_learning_curves_bool = True)

# train and test the model accordingly
cnn_forecast, test, train = cnn_model.fit()
```

### GRU

```python
# initialise the model
gru_model = gru(df = df, target = target, test_size = test_size_prct, plot_learning_curves_bool = False)

# train and test the model accordingly
gru_forecast, test, train = gru_model.fit()
```

### VAR

```python
#Initialise the class
var_model = var(df = df, target = target , test_size = test_size_prct, max_lag= 5)
#Forecast the time series data
var_forecast, test, train=var_model.fit()
```
